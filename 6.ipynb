```
import sqlite3

connection = sqlite3.connect('library.db')

cursor = connection.cursor()

create = '''
CREATE TABLE IF NOT EXISTS book (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  title TEXT NOT NULL,
  author TEXT NOT NULL,
  year_published INTEGER,
  genre TEXT
);
'''

cursor.execute(create)
connection.commit()
connection.close()
```
