```python
def words(text):
  counts = {}
  words = text.split()
  for word in words:
    norm = word.lower().strip('.,!?;:"\'()[]{}')
    if norm:
      if norm in counts:
        counts[norm] += 1
      else:
        counts[norm] = 1
  return counts

text = "qwe qwe ew we ew er re re"
word = words(text)
print(word)
```

```python
def invert(original_dict):
  inv = {}
  for key, value in original_dict.items():
    if value not in inv:
      inv[value] = key
    else:
      if isinstance(inv[value], list):
        inv[value].append(key)
      else:
        inv[value] = [inv[value], key]
  return inv
q = {'a': 1, 'b': 2, 'c': 3}
print(invert(q))
```

```python
def merge(a, b):
  merged = {}
  for key, value in a.items():
    merged[key] = value
    for key, value in b.items():
      if key in merged:
        merged[key] += value
      else:
        merged[key] = value
  return merged

a = {'a': 1, 'b': 2}
b = {'c': 12, 'd': 16}
print(merge(a, b))
```

```python
def filter(original_dict, t):
  return {k: v for k, v in original_dict.items() if v > t}

n = {'a': 10, 'b': 5, 'c': 15, 'd': 3, 'e': 12}
t = 8

fn = filter(n, t)
print(fn)
```

```python
def maxx(d):
  if not d:
    return None
  return max(d, key=d.get)

data = {'a': 10, 'b': 5, 'c': 15, 'd': 3, 'e': 12}
max_key = maxx(data)
print(f"   : {max_key}")
```

```python
from collections import Counter
def uni(q):
  count = Counter(q)
  return [item for item in q if count[item] == 1]

n = [1, 1, 2, 3, 3, 3, 4, 5, 5, 6, 6, 7, 7]
u = uni(n)
print(u)
```

```python
def dell(q):
  lis = set()
  u = []
  for item in q:
      if item not in lis:
          u.append(item)
          lis.add(item)
  return u
n = [1, 1, 2, 3, 3, 3, 4, 5, 5, 6, 6, 7, 7]
d = dell(n)
print(d)
```

```python
def un(set1, set2):
  return set1 | set2
def d(set1, set2):
  diff1 = set1 - set2
  diff2 = set2 - set1
  return un(diff1, diff2)
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
q = d(a, b)
print(q)
```
