```python
def words(text):
  counts = {}
  words = text.split()

  for word in words:
    norm = word.lower().strip('.,!?;:"\'()[]{}')
    if norm:
      if norm in counts:
        counts[norm] += 1
      else:
        counts[norm] = 1

  return counts


text = "qwe qwe ew we ew er re re"
word = words(text)
print(word)
```
