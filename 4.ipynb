```python
def words(text):
  counts = {}
  words = text.split()
  for word in words:
    norm = word.lower().strip('.,!?;:"\'()[]{}')
    if norm:
      if norm in counts:
        counts[norm] += 1
      else:
        counts[norm] = 1
  return counts

text = "qwe qwe ew we ew er re re"
word = words(text)
print(word)
```

```python
def invert(original_dict):
  inv = {}
  for key, value in original_dict.items():
    if value not in inv:
      inv[value] = key
    else:
      if isinstance(inv[value], list):
        inv[value].append(key)
      else:
        inv[value] = [inv[value], key]
  return inv
q = {'a': 1, 'b': 2, 'c': 3}
print(invert(q))
```

```python
def merge(a, b):
    merged = {}
    for key, value in a.items():
      merged[key] = value
      for key, value in b.items():
        if key in merged:
          merged[key] += value
        else:
          merged[key] = value
    return merged

a = {'a': 1, 'b': 2}
b = {'c': 12, 'd': 16}
print(merge(a, b))
```
