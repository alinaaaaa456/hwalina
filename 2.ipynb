```python
class H:
  def __init__(self, person):
    self.__bit = 60
    self.__w = True
    self.__owner = person
  def beats(self):
    if self.__w:
      print(f"Сердце бьётся с частотой {self.__bit} ударов в минуту")
      return True
    print(f"Сердце не бьётся")
    return False
  def stop(self):
    self.__w=False
    self.__bit=0
    print(f"Сердце не бьётся")
  def change_bit(self, x):
    if self.__w and 20 <= x <= 200:
      self.__bit = x
      print(f"Сердце бьётся с частотой {self.__bit} ударов в минуту")
      return True
    else:
      print("Сердце не бьётся")
      return False
class Person:
  def __init__(self, name):
    self.__name= name
    self.__h=H(self)
  @property
  def name(self):
    return self.__name
  def alive(self):
    if self.__h.beats():
      print(f"{self.name} жив")
      return True
    print(f"{self.name} мёртв")
    return False
  def stop_beats(self):
    self.__h.stop()
  def change_h_bit(self, rate):
    return self.__h.change_bit(rate)
name=Person("Alina")
name.change_h_bit(10)
name.change_h_bit(70)
name.alive()
```

```python
class Player:
  def __init__(self, name, position):
    self.name = name
    self.position = position
    def __str__(self):
      return f"{self.name} ({self.position})"
class Team:
  def __init__(self, name):
    self.name = name
    self.players = []
  def add_player(self, player):
    if isinstance(player, Player):
      self.players.append(player)
      print(f"{player.name} добавлен в команду {self.name}")
    else:
      print("ошибка")
  def remove_player(self, player_name):
    for player in self.players:
      if player.name == player_name:
        self.players.remove(player)
        print(f"{player_name} удален из команды {self.name}")
        return
      print(f"Игрок {player_name} не найден в команде {self.name}")

  def display_team(self):
    print(f"Состав команды {self.name}:")
    for player in self.players:
      print(f"- {player}")

p1 = Player("Данил", "Нападающий")
p2 = Player("Руслан", "Защитник")
p3 = Player("Камиль", "Вратарь")
p4 = Player("Алексей", "Нападающий")
t1 = Team("Бекэнд")
t2 = Team("Фронтенд")
t1.add_player(p1)
t1.add_player(p2)
t2.add_player(p3)
t2.add_player(p4)

t1.display_team()

t2.display_team()

t1.remove_player("Данил")
t1.display_team()

t1.display_team()
```
