```python
def matrix_norm(matrix):
  max_abs = 0
  for row in matrix:
    for i in row:
      current_abs = abs(i)
      if current_abs > max_abs:
        max_abs = current_abs
  return max_abs


def disp(matrix):
  for i in matrix:
    print(i)

A = [
    [5, -2, 3],
    [1, 0, -4],
    [-6, 2, 7]
]

B = [
    [-1, 2, 0],
    [3, -5, 1],
    [2, 4, -3]
]

C = [
    [6, -1, 2],
    [0, 3, -2],
    [1, -4, 5]
]

norm_A = matrix_norm(A)
norm_B = matrix_norm(B)
norm_C = matrix_norm(C)

min_norm = min(norm_A, norm_B, norm_C)

if min_norm == norm_A:
    print(" A   :", min_norm)
    print("A =")
    disp(A)
elif min_norm == norm_B:
    print(" B   :", min_norm)
    print("B =")
    disp(B)

else:
    print(" C   :", min_norm)
    print("C =")
    disp(C)
```
