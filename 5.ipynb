```python
s='qwe qwe wert ty.'

if s.endswith('.'):
  s = s[:-1]
words = s.split()

if not words:
  print("Текст не содержит слов")
else:
  lengths = [len(word) for word in words]

  min_len = min(lengths)
  max_len = max(lengths)

  print(f"Длина самого короткого слова: {min_len}")
  print(f"Длина самого длинного слова: {max_len}")
```

```python
import random
with open('file.txt', 'w') as f:
  for i in range(20):
    num = random.randint(1, 10)
    f.write(f"{num}\n")
numbers = set()

with open('file.txt', 'r') as f, open('file2.txt', 'w') as g:
  for line in f:
    num = int(line.strip())
    if num not in numbers:
      numbers.add(num)
      g.write(f"{num}\n")

with open('file.txt', 'r') as f:
  print("Содержимое файла f:")
  for line in f:
    print(line.strip(), end=' ')

with open('file2.txt', 'r') as g:
  print("\nСодержимое файла g (без повторений):")
  for line in g:
    print(line.strip(), end=' ')
```

```python
with open('Bibl.txt', 'r') as file:
  books = file.readlines()

found_books = []
for book in books:
  parts = book.strip().split()
  if len(parts) >= 3:
    author = parts[0]
    title = ' '.join(parts[1:-1])
    year = parts[-1]

    if "Информатика" in title:
      found_books.append((author, year))

if not found_books:
  print("Книг с названием 'Информатика' не найдено")
else:
  print("Найдены следующие книги с названием 'Информатика':")
  for author, year in found_books:
      print(f"Автор: {author}, Год издания: {year}")
```

```python
from collections import Counter
import re

def frequency_analysis(file_path):
  with open('Bibl.txt', 'r') as file:text = file.read()

  letters = re.findall(r'[А-ЯЁ]', text)

  letter_counts = Counter(letters)
  total_letters = sum(letter_counts.values())

  frequency = {letter: (count / total_letters) * 100 for letter, count in letter_counts.items()}

  sorted_frequency = sorted(frequency.items(), key=lambda x: x[1], reverse=True)

  return sorted_frequency

file_path = 'text.txt'
result = frequency_analysis(file_path)

for letter, percent in result:
    print(f"{letter}: {percent:.2f}%")
```
